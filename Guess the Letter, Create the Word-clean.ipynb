**Create a Letter Guessing Game**

Create a program that allows a user 7 chances to guess the correct word. If they do not guess the word after 7 tries, the player loses and the program will print the correct word. Otherwise, the play wins and the game.

**How to build the program**
* When you start the game, you will need to select a random word from a list of at least 10 words (You have full control over which words you want to use for you program). This will be your secret word. Your secret word will be represented in the program as a group of underscores. For as long as the word is, you should also have that many underscores. 

* Once the word is selected, your game will commence. Perform a Google search to figure out how to select a random word from a list using Python.<br>
**Hint** There's package you can import into your application that does this for you.

* The end user will have a total of 7 chances to guess the correct letter from the secret word. If the end user makes 7 incorrect guesses, the game will end.

* As you guess the correct letters, the letters you have guess will then take place of the underscores that letter represents. <br>
**For Example**: If your secret word is 'watermelon' and  so far you have guessed the letters 'a' and 'e', the word you're trying to guess will appear as follows: _ a _ e _ _ e _ _ _.<br>
**Keep in mind** that if you guess a letter that appears more than once in your secret word, make sure that the letter is populated anywhere that letter would be.

```python
import random
from IPython.display import clear_output as co

words = {
    'fruits':['mango', 'orange', 'kiwi' , 'dragonfruit', 'peach', 'apple', 'watermelon', 'cucumber', 'jackfruit', 'durian'],
    'animals' : ['springbok', 'anteater', 'hippopotamus', 'elephant', 'hyena', 'capybara', 'vulture', 'warthog', 'meerkat', 'aardvark' ],
    'flowers' : ['chrysanthemum', 'freesia', 'hyacinth', 'gladiolus', 'anemone', 'sunflower', 'daffodil', 'carnation', 'peony','rafflesia'],
}


class WordBank:
    def __init__(self, words_dict = words):
        """Trying to instantiate words dictionary"""
        for key in words_dict:
            setattr(self, key, words_dict[key])
        
    def show_categories(self, cat):
        """Display all categories"""
        for key in cat.keys():
            print(key.capitalize())

    def get_random_word(self, category):
        """User chooses category and gets random word"""
        #getattr =  gets attribute of category
        word = random.choice(getattr(self, category))
        return word.upper()
    
class Guesses:
    
    def __init__(self, answer):
        """instantiate guess and wrong answer list"""
        #all the letters guessed
        self.guessed_li = []
        #all wrong letter guessed
        self.wrong_li = []
        self.answer = answer
        self.blank_answer = None
        
    def create_blanks(self):
        self.blank_answer =  ['_' for item in range(len(self.answer))]
        print(self.blank_answer)
#         for i in self.blank_answer:
#             print()
    
    
    #print(random_word)

    def is_it_worth_it(self, the_guess):
        """is it worth running comparison if not real input"""
        #if more than one letter is submitted
        if len(the_guess) != 1:
            print('Please guess one letter at a time.')
            return False
        #if the same letter is guessed twice
        if the_guess in self.guessed_li:
            print('You have already guessed this letter.')
            return False
        #if a character is chosen instead of a letter:
        if not the_guess.isalpha():
            print("Please choose a LETTER")
            return False
        return True

    def guess(self, the_guess):
        """Checks if users letter is in word"""
#         print(self.answer)
#         print(the_guess)
        index = 0
        found = False
        for l in self.answer:
#             print(l)
#             print(type(l))
#             print(type(the_guess))
            if l == the_guess:
                self.blank_answer[index] = the_guess  
                found = True
            index += 1
        if found == False:
             self.wrong_li.append(the_guess)
#         print(found)
        return found

#         self.guessed_li.append(the_guess)
#         for x in self.answer:
#             if x in self.blank_answer:
#                 print(' ' + x + '', end='')
#             elif x not in self.blank_answer:
#                 print(' __', end='')
#         if the_guess in self.answer:
#             new_r_w = self.answer.index(the_guess)
#             #all blanks until index + the_guess + all the blanks following
           
#             new_blank = self.blank_word[:new_r_w] + the_guess + self.blank_word[new_r_w + 1:]
#             self.blank_answer = new_blank
#             return True

    
    def get_blank_word(self):
        return self.blank_answer

    def wrong_ans(self):
        """Counts the wrong answers"""
        attempts = len(self.wrong_li)
        return attempts 

    def guess_word(self, guess_w):
        """If you know the word"""
        return guess_w == self.answer
                   

class Game:
    
    def show_instructions(self):
        print("""It's time to play Hangman! 
A word will be randomly generated. All you have to do is guess the letters or if you're feeling bold, guess the word.\n""")

    
    
    def run(self):
        """Playing the game"""
        """Instantiating dictionary into WordBank"""
        wordbank = WordBank(words)
        
        done = False
        
        Game.show_instructions(self) 
        
        wordbank.show_categories(words)
        category = input("Please choose a category: \n")
        secret_word = wordbank.get_random_word(category)
        print(secret_word)
        guesses = Guesses(secret_word)                     
#         print(guesses.get_blank_word())
        while not done:
            #succeeding = True

            choice = input("Would you like to guess the 'letter' or the 'word'? \n").lower()
            guesses.create_blanks()
            if choice == 'letter' or choice == 'l':
                g_letter = input("Guess a letter: \n").upper()
                co()
                worth_it = guesses.is_it_worth_it(g_letter)
            if worth_it == True:
                compared = guesses.guess(g_letter)
                if compared == True:
                    good_job = guesses.get_blank_word()
                    print(good_job)
                else:
                    print("That's wrong.")
                    print(f"You have been incorrect {guesses.wrong_ans()} time(s).")
                    if guesses.wrong_ans() >= 7:
                        done = True
                        print("HAHA Game over!")
                        break
            if choice == 'word' or choice == 'w':
                the_word = input("What do you think the word is? If you get it wrong, you lose. \n").upper()
                attempt_word = guesses.guess_word(the_word)
                if attempt_word == False:
                    done = True
                    print("HAHA Game over!")
                    break
                else:
                    print('YOU WIN')
                            
    #input guess letter or word
    #input guess letter
    #input guess word
    #compare against hidden word

g = Game()
g.run()
```
